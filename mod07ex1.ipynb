````python
df_columns = list(df.columns)
metadados_dict = { 'coluna': df_columns, 'tipo': list(df.dtypes) }
metadados_df = pd.DataFrame(metadados_dict)
metadados_df
````

````python
categorias_unicas_linha = []
for col in df_columns:
    categorias_unicas_linha.append(len(df[col].unique()))
    
metadados_df['qtd_categorias'] = categorias_unicas_linha
metadados_df
````

````python
metadados_encoded = pd.get_dummies(metadados_df, columns=['coluna', 'tipo'], drop_first=True)
metadados_encoded.head()
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
df.dtypes
````

````python
encoded_df = pd.get_dummies(df, columns=['tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia'], drop_first=True)
encoded_df.head()
````

````python
encoded_df.sexo = encoded_df.sexo.map({'F': 1, 'M': 0})
encoded_df.posse_de_veiculo = encoded_df.posse_de_veiculo.map({'Y': 1, 'N': 0})
encoded_df.posse_de_imovel = encoded_df.posse_de_imovel.map({'Y': 1, 'N': 0})
encoded_df.head()
````

````python
ct = pd.crosstab(encoded_df['possui_email'], encoded_df['mau'])
ct
````

````python
ct = pd.crosstab(encoded_df['posse_de_veiculo'], encoded_df['mau'])
ct
````

A variável mais poderosa para prever inadimplência somente com base nas duas tabelas cruzadas é ```possui_email```, porque mais de 83% dos clientes não possuem email, enquanto apenas menos de 17% possuem. É uma diferença de comportamento grande o suficiente para superar a ```posse_de_veiculo```, pois cerca de 63% da base de clientes inadimplentes não possuem veículos, e 37% possuem.
